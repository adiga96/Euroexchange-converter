```python
# Establish the connection to the particular API
conn = http.client.HTTPSConnection("currency-converter5.p.rapidapi.com")

headers = {
    'x-rapidapi-host': "currency-converter5.p.rapidapi.com",
    'x-rapidapi-key': RAPIDAPIKEY
    }
```

```python
# First we will find the List of available currencies:
conn.request("GET", "/currency/list", headers=headers)

# Get the data response and order the format
resp  = conn.getresponse()
data = resp.read().decode("utf-8")

#obtain a dictionary of the list of currencies
print(type(data))
print(data)
currency = json.loads(data)
```

```python
# Next, we will check the currency converter on the base of 1 euro
conn.request("GET", "/currency/convert", headers=headers)

resp = conn.getresponse()
data = resp.read().decode("utf-8")
convert = json.loads(data)
convert
```

```python
# The rate value for 1 euro is visualized in the following dictionary, updated to the current date
for key in convert.keys():
    print("key =", key)
    print(convert[key])
    print("\n")
```

```python
# change to list visualization
convert['rates']
```

```python
# panda dataframe convert rate 
convert_df= pd.DataFrame(convert['rates'])
convert_df
```

```python
# get the Historical Currency Rates specifying the year - month - day
conn.request("GET", "/currency/historical/2019-12-10", headers=headers)

resp = conn.getresponse()
data = resp.read().decode("utf-8")
historical = json.loads(data)
historical
```
